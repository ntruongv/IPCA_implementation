```python
# Augment characteristics 
Zts = np.array(Zts_save, copy=True)
# Zts_mean = np.mean(Zts, axis=2)
# Zts = np.concatenate((np.ones((Zts.shape[0],1,Zts.shape[2])), 
#                       np.repeat(np.expand_dims(Zts_mean,axis=2), Zts.shape[2], axis=2), 
#                       Zts - Zts_mean.reshape(Zts.shape[0],Zts.shape[1],1)), axis = 1)

# normalize rank 
Zts = np.argsort(Zts, axis=0)/Zts.shape[0]-.5
print("Zts.shape (N,L,T)", Zts.shape)
print("np.amax(Zts)", np.amax(Zts))
print("np.amin(Zts)", np.amin(Zts))

Zt_cov = np.zeros((Zts.shape[1],Zts.shape[1],Zts.shape[2]))
x_t = np.zeros((Zts.shape[1],Zts.shape[2]))
for i_t in range(Zts.shape[2]):
    Zt = Zts[:,:,i_t]
    Zt_cov[:,:,i_t] = (Zt.T).dot(Zt)
    if i_t<Zts.shape[2]-1:
        x_t[:,i_t] = (Zt.T).dot(rs[:,i_t+1])
```

# Model specification 
$$r_{i,t+1}=\alpha_{i,t}+\beta_{i,t}f_{t+1}+\epsilon_{i,t+1}$$
$$\alpha_{i,t}=z_{i,t}^\prime\Gamma_\alpha+\nu_{\alpha,i,t}, \beta_{i,t}=z_{i,t}^\prime\Gamma_\beta+\nu_{\beta,i,t}$$
where $f$ is factor vector; $z$ is characteristics factor. 

# Implementation
## Restricted model $\Gamma_\alpha=0$
Given information up to time $t$, the return vector (size $N\times 1$) can be written as 
$$r_{t+1}=Z_t\Gamma_\beta f_{t+1}+\epsilon_{t+1}^*$$
where $r_{t+1}$ is the return vector; $Z_t$ is $N\times L$ matrix of characteristics; $\Gamma_\beta$ is $L\times K$ matrix. 

Alternating least squares: 
- If $\Gamma_\beta$ is known. Need to solve $r_{t+1}=Z_t\Gamma_\beta f_{t+1},\forall t$. Use $\hat f_{t+1}=\left(\Gamma_\beta^TZ_t^TZ_t\Gamma_\beta\right)^{-1}\Gamma_\beta^TZ_t^Tr_{t+1}$
- If $f_t$ is known. Need to jointly solve $r_{t+1}=Z_t\otimes f_{t+1}^T\text{vec}\left(\Gamma_\beta^T\right),\forall t$. Use $\text{vec}\left(\Gamma_\beta^T\right)=\left(\sum_{t=1}^{T-1}\left[Z_t\otimes f_{t+1}^T\right]^T\left[Z_t\otimes f_{t+1}^T\right]\right)^{-1}\left(\sum_{t=1}^{T-1}\left[Z_t\otimes f_{t+1}^T\right]^Tr_{t+1}\right)$

where vectorization in $\text{vec}\left(\Gamma_\beta^T\right)$ is defined as `reshape` by column in Python. 

```python
K = 5
L = Zts.shape[1]
N = Zts.shape[0]
print("K,L,N",K,L,N)
# pca = PCA(n_components = K)
mat = np.zeros((L,L))
for i_t in range(T-1): 
    Zt = Zts[:,:,i_t]
    mat0 = Zt.T.dot(rs[:,i_t+1])
    mat += mat0.dot(mat0.T)/(Zt.shape[0])**2
# pca.fit(mat)
# Gamma_b_0 = pca.components_.T
_, eig_vec = np.linalg.eigh(mat)
Gamma_b_0 = eig_vec[:,np.arange(-1,-K-1,-1)]

plt.figure(); plt.hist(Gamma_b_0.flatten(), bins=(Gamma_b_0.shape[0]*Gamma_b_0.shape[1])//15);
print(Gamma_b_0.shape)
Gamma_b_init = np.array(Gamma_b_0, copy=True)
```

```python
np.random.seed(4134132)
K = 5
L = Zts.shape[1]
N = Zts.shape[0]
# initialization 
Gamma_b_0 = np.random.rand(L,K)
# Gamma_b_0 = np.array(Gamma_b_init, copy=True)
Gamma_b_1 = np.array(Gamma_b_0, copy=True)
F_0 = np.random.rand(K,T)
F_1 = np.array(F_0, copy=True)
iter = 0
tol = 1e-5
start_time = time.time()
max_iter = 1001 
max_abs_delta_F = np.zeros(max_iter)
max_abs_delta_Gamma_b = np.zeros(max_iter)
while iter < max_iter:
# while iter < 1:
    # update F
    for i_t in range(T-1): 
        #Zt = Zts[:,:,i_t]
        mat = Gamma_b_0.T.dot(Zt_cov[:,:,i_t]).dot(Gamma_b_0)
        F_1[:,i_t+1] = np.linalg.inv(mat).dot(Gamma_b_0.T).dot(x_t[:,i_t]) 
    # update Gamma_b
    mat_1 = np.zeros((L*K,L*K))
    mat_2 = np.zeros(L*K)
    for i_t in range(T-1):
        #Zt = Zts[:,:,i_t]
        #mat_0 = np.kron(Zt, F_1[:,i_t+1].T)
        cur_F = F_1[:,i_t+1].reshape(-1,1)
        mat_1 += np.kron(Zt_cov[:,:,i_t], (cur_F.dot(cur_F.T)) ) # mat_0.T.dot(mat_0)
        mat_2 += np.kron(x_t[:,i_t], cur_F.flatten()) #mat_0.T.dot(rs[:,i_t+1])
    Gamma_b_1 = np.linalg.inv(mat_1).dot(mat_2).reshape(L,K)
    # normalization 
    U,S,V = np.linalg.svd(Gamma_b_1, full_matrices=False)
    Gamma_b_1 = Gamma_b_1.dot(V.T).dot(np.diag(1/S))

    max_abs_delta_F[iter] = np.sum(np.abs(F_0-F_1)) / (T * K)
    max_abs_delta_Gamma_b[iter] = np.sum(np.abs(Gamma_b_0-Gamma_b_1)) / (L * K)
    # max_abs_delta_F[iter] = np.amax(np.abs(F_0-F_1))
    # max_abs_delta_Gamma_b[iter] = np.amax(np.abs(Gamma_b_0-Gamma_b_1))
    
    if max_abs_delta_F[iter] <= tol and max_abs_delta_Gamma_b[iter] <= tol:  
        print('iter', iter, '\tmax_abs_delta_F', max_abs_delta_F[iter], 
              '\tmax_abs_delta_Gamma_b', max_abs_delta_Gamma_b[iter])
        break
    if not iter % 100: 
        print('iter', iter, '\tmax_abs_delta_F', max_abs_delta_F[iter], 
              '\tmax_abs_delta_Gamma_b', max_abs_delta_Gamma_b[iter])
    Gamma_b_0 = np.array(Gamma_b_1, copy=True)
    F_0 = np.array(F_1, copy=True)

    iter += 1 
end_time = time.time()
print("Elapsed time", end_time - start_time)
```

## Unrestricted model $\Gamma_\alpha\neq0$
Given information up to time $t$, the return vector (size $N\times 1$) can be written as 
$$r_{t+1}=Z_t\Gamma_\alpha+Z_t\Gamma_\beta f_{t+1}+\epsilon_{t+1}^*$$
where 
- $r_{t+1}$ is the return vector
- $Z_t$ is $N\times L$ matrix of characteristics
- $\Gamma_\alpha$ is $L\times 1$ vector
- $\Gamma_\beta$ is $L\times K$ matrix 

Alternating least squares: 
- If $\Gamma_\alpha,\Gamma_\beta$ are known. Use $\hat f_{t+1}=\left(\Gamma_\beta^TZ_t^TZ_t\Gamma_\beta\right)^{-1}\Gamma_\beta^TZ_t^T(r_{t+1}-Z_t\Gamma_\alpha)$
- If $f_t$ is known. Need to jointly solve $r_{t+1}=Z_t\otimes f_{t+1}^T\text{vec}\left(\Gamma_\beta^T\right),\forall t$. Use $\text{vec}\left(\Gamma_\beta^T\right)=\left(\sum_{t=1}^{T-1}\left[Z_t\otimes \tilde f_{t+1}^T\right]^T\left[Z_t\otimes \tilde f_{t+1}^T\right]\right)^{-1}\left(\sum_{t=1}^{T-1}\left[Z_t\otimes \tilde f_{t+1}^T\right]^Tr_{t+1}\right)$

where vectorization in $\text{vec}\left(\Gamma_\beta^T\right)$ is defined as `reshape` by column in Python. 

```python
# DID NOT IMPLEMENT 
np.random.seed(4134132)
K = 5
L = Zts.shape[1]
N = Zts.shape[0]
# initialization 
Gamma_0 = np.random.rand(L, 1+K)
Gamma_1 = np.array(Gamma_0, copy=True)
Gamma_a = Gamma_0[:,0]
Gamma_b = Gamma_0[:,1:]
F_0 = np.random.rand(K, T)
F_1 = np.array(F_0, copy=True)
tol = 1e-6
start_time = time.time()
max_iter = 1001 
max_abs_delta_F = np.zeros(max_iter)
max_abs_delta_Gamma = np.zeros(max_iter)
for iter in range(max_iter):
    # update F
    for i_t in range(T-1): 
        Zt = Zts[:,:,i_t]
        mat = Gamma_b.T.dot(Zt.T).dot(Zt).dot(Gamma_b)
        F_1[:,i_t+1] = np.linalg.inv(mat).dot(Gamma_b.T).dot(Zt.T).dot(rs[:,i_t+1]-Zt.dot(Gamma_a)) 
    # update Gamma_b
    tilde_F_1 = np.concatenate((np.ones((1,T)), F_1), axis = 0) # (1+K, T)
    mat_1 = np.zeros((L*(1+K),L*(1+K)))
    mat_2 = np.zeros(L*(1+K))
    for i_t in range(T-1):
        Zt = Zts[:,:,i_t]
        mat_0 = np.kron(Zt, tilde_F_1[:,i_t+1].T)
        mat_1 += mat_0.T.dot(mat_0)
        mat_2 += mat_0.T.dot(rs[:,i_t+1])
    Gamma_1 = np.linalg.inv(mat_1).dot(mat_2).reshape(L,1+K)
    # normalization 
    U,S,V = np.linalg.svd(Gamma_1[:,1:], full_matrices=False)
    Gamma_b = Gamma_1[:,1:].dot(V.T).dot(np.diag(1/S))
    Gamma_1[:,1:] = Gamma_b 
    
    Gamma_a = Gamma_1[:,0]
    Gamma_a = Gamma_a-Gamma_b.dot(np.linalg.inv(Gamma_b.T.dot(Gamma_b))).dot(Gamma_b.T).dot(Gamma_a)
    Gamma_1[:,0] = Gamma_a

    max_abs_delta_F[iter] = np.sum(np.abs(F_0[:,1:]-F_1[:,1:])) / ((T-1) * K)
    max_abs_delta_Gamma[iter] = np.sum(np.abs(Gamma_0-Gamma_1)) / (L * (1+K))
    if max_abs_delta_F[iter] <= tol and max_abs_delta_Gamma[iter] <= tol:  
        print('iter', iter, '\tmax_abs_delta_F', max_abs_delta_F[iter], 
              '\tmax_abs_delta_Gamma', max_abs_delta_Gamma[iter])
        break
    if not iter % 100: 
        print('iter', iter, '\tmax_abs_delta_F', max_abs_delta_F[iter], 
              '\tmax_abs_delta_Gamma', max_abs_delta_Gamma[iter])
    Gamma_0 = np.array(Gamma_1, copy=True)
    F_0 = np.array(F_1, copy=True)

    iter += 1 
end_time = time.time()
print("Elapsed time", end_time - start_time)
```

```python
plt.figure(); 
plot_mat = max_abs_delta_Gamma[max_abs_delta_Gamma>tol/2]
plt.plot(np.arange(plot_mat.shape[0]), plot_mat, '.');
plot_mat = max_abs_delta_F[max_abs_delta_F>tol/2]
plt.plot(np.arange(plot_mat.shape[0]), plot_mat, '.');
plt.yscale('log'); plt.xlabel('Iteration'); plt.ylabel('Costs'); 
plt.legend(('Delta Gamma', 'Delta F')); 
plt.tight_layout();
plt.savefig('./figures/Unrestricted_costs_vs_iteration.pdf');
```
